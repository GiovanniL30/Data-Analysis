```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('clean.csv')
```

```python
result = pd.DataFrame({
    "Mean": df['yearly_med_salary'].mean(), 
    "Median": df['yearly_med_salary'].median(), 
    "Mode": df['yearly_med_salary'].mode(), 
    "Midrange": df['yearly_med_salary'].max() - df['yearly_med_salary'].min() / 2
    })

result
```

```python
result = pd.DataFrame({
    "Variance": ["{:.2f}".format(df['yearly_med_salary'].var())],
    "Standard Deviation": ["{:.2f}".format(df['yearly_med_salary'].std())]
})

result
```

```python
# Calculate the mean, standard deviation, and variance
mean_salary = df['yearly_med_salary'].mean()
std_dev_salary = df['yearly_med_salary'].std()
variance_salary = df['yearly_med_salary'].var()

# Plotting the KDE without color fill
plt.figure(figsize=(14, 8))
sns.kdeplot(df['yearly_med_salary'], color='blue', shade=False, linewidth=2)

# Add lines for mean, mean ± std dev
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean_salary:.2f}")
plt.axvline(mean_salary + std_dev_salary, color='green', linestyle='dashed', linewidth=2, label=f"k1: {mean_salary + std_dev_salary:.2f}")
plt.axvline(mean_salary - std_dev_salary, color='green', linestyle='dashed', linewidth=2, label=f"k1: {mean_salary - std_dev_salary:.2f}")


# Plotting details
plt.xlabel("Salary (USD)", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.title("Salary Distribution", fontsize=16)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks([])
plt.yticks([])
plt.show()


```
