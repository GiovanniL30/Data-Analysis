```python
# Author: Kevin Roy Maglaqui
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('job_posting_clean.csv')
```

# Which job postings receive the most applications?

```python
#list all the number of applications
#df.applies.to_string(index=False)
most_applications = df.loc[df['applies'].idxmax()]

# Graph of the most applied jobs
top_jobs = df.nlargest(10, 'applies')
plt.figure(figsize=(12,6))
sns.barplot(x='applies', y='title', data=top_jobs, palette='viridis')
plt.title("Most applied Jobs (July 2023-November 2023)")
plt.xlabel("Applications")
plt.ylabel("Job Title")
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

# Is there a correlation of jobs between salary and job views/applications?

```python
# Group by job title and calculate the average salary, views, and applications
grouped_df = df.groupby('title')[['yearly_med_salary', 'applies']].mean().reset_index()

correlation_applies_salary = grouped_df['yearly_med_salary'].corr(grouped_df['applies'])

print(f"Correlation between Salary and Applications: {correlation_applies_salary:.2f}")

plt.scatter(grouped_df['applies'], grouped_df['yearly_med_salary'], alpha=0.6, color="royalblue")
plt.title('Salary vs Applications by Job Title')
plt.xlabel('Average Applications')
plt.ylabel('Average Salary (USD)')
plt.grid(True)

plt.show()
```
